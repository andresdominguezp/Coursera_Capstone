### Background

**Tegucigalpa**

Honduras is a small developing country located in Central America. It's capital, Tegucigalpa, is Honduras' largest and most populous city as well as the nation's political and administrative center.

**Why a retail recommender system in Tegucigalpa?**

Being the most populous city in Honduras, Tegucigalpa has a great potential to open up retail stores. 

People here thrive day and night for opportunities to generate incomes, and opening retails stores is one of the options. However, some retails are less successful than others because of their location. I have seen some of my friends' retail stores close o relocate because they did not have enough traffic where they opened it up.

With this recommender system, I would like to solve that problem and help people to best pick a good location for their new retail store.

**What is a retail store?**

According to Money Matters: All Management Articles, "A retailer or retail store is a business enterprise whose primary source of selling comes from retailing.

"Retailing includes all the activities involved in selling of goods or services directly to final consumer for personal, non-business use" – Philip Kotler."

**Things to consider when opening a retail store**

Of course, there are many different factors to take into consideration when opening a retail store, however, we will focus on the **location** of the retail store. This factor greatly impacts the success of a retail store.

According to Shari Waters, "retailers would want to be located where there are many shoppers but only if those shoppers meet the definition of their target market. Small retail stores may benefit from the traffic generated by nearby larger stores. There are several aspects retailers should consider along these lines:

+ How many people walk or drive past the location?
+ How well is the area served by public transportation?
+ Can customers and delivery trucks easily get in and out of the parking lot?
+ Is there adequate parking?

Other thing to take in mind is if the types of businesses nearby are compatible with your store. For example, a high-end fashion boutique may not be successful next door to a discount variety store. Position it next to a nail or hair salon, which tend to draw the same demographic of customers, to more optimal results."

With this information, we can start visualizing the data that we will have to take into consideration when building our recommendation system. With this project, I am looking to facilitate the location factor.
